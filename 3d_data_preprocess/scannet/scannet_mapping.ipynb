```python
import pandas as pd
import os
import os.path as osp

scannet_path = '/media/navigation/easystore/Original_dataset/scannet/data'

scannet_mappings = pd.read_csv(osp.join(scannet_path, 'scannetv2-labels.combined.tsv'), sep='\t')
with open(osp.join(scannet_path, 'nyu_label.csv'), 'r') as f:
    nyu_classes = [line.strip('\n').strip(',') for line in f.readlines()]
nyu_class_to_id = {nyu_class: i+1 for i, nyu_class in enumerate(nyu_classes)}
```

```python
nyuid_to_scannet_id = {}
scannet_mappings.nyuClass = scannet_mappings.nyuClass.fillna('unknown')
for id, nyuclass in enumerate(nyu_classes):
    scannet_rows = scannet_mappings[scannet_mappings.nyuClass == nyuclass]
    scannet_ids = scannet_rows.index
    if len(scannet_ids) > 0:
        nyuid_to_scannet_id[id+1] = {
            'scannet_id': int(scannet_ids[0]), 
            'scannet_class': scannet_rows.raw_category.iloc[0],
            'nyu_class': nyuclass
            }
    else:
        nyuid_to_scannet_id[id+1] = {
            'scannet_id': int(5), 
            'scannet_class': 'object',
            'nyu_class': nyuclass
            }
        
    
```

```python
import json
with open('nyu_id_to_scannet_id.json', 'w') as f:
    json.dump(nyuid_to_scannet_id, f)
```
